```python

import socket,cv2,threading
import numpy as np

s = socket.socket()
s.bind(("", 1234))
s.listen()
session, address = s.accept()
print(session.recv(10000000)) 
camera = cv2.VideoCapture("http://192.168.198.105:8080/video")

def sender():
    while True:
        status, photo = camera.read()
        photo = cv2.resize(photo, (640, 480))
        print(photo.shape)
        if status:
            session.send(np.ndarray.tobytes(photo))
        else:
            print("frame not found")

def receiver():
    frames_error = 0
    print("Entered")
    while True:
        data = session.recv(100000000)
        if(data == b'finished'):
            print("Finished")
            camera.release()
            session.close()
        else:
            photo =  np.frombuffer(data, dtype=np.uint8)
            if len(photo) == 640*480*3:
                cv2.imshow('From Client side', photo.reshape(480, 640, 3))
                if cv2.waitKey(1) == 13:
                    session.send(b'finished')
                    camera.release()
                    cv2.destroyAllWindows()
                    break
            else:
                print(f"Unavailable {frames_error} frames" )
                
threading.Thread(target=sender).start()
threading.Thread(target=receiver).start()
```
