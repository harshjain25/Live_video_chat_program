```python
import socket ,cv2,threading
import numpy as np

s= socket.socket()
s.bind(("",8081))

sender_ip=input("Enter the Sender Ip : ")
sender_port=int(input("Enter the Sender port : "))
s.connect((sender_ip,sender_port))

camera = cv2.VideoCapture(0)


def receiver():
    frames_error = 0
    print("Entered")
    while True:
        data = s.recv(10000000000)
        if(data == b'finished'):
            print("Finished")
            camera.release()
            s.close()
        else:  
            photo =  np.frombuffer(data, dtype=np.uint8)
            if len(photo) == 640*480*3: 
                cv2.imshow('From Server', photo.reshape(480, 640, 3))
                if cv2.waitKey(1) == 13: 
                    s.send(b'finished')
                    camera.release()
                    cv2.destroyAllWindows()
                    break
            else:
                print(f"Unavailable  {frames_error} frames" )
 


def sender():
    while True: 
        status, photo = camera.read()
        photo = cv2.resize(photo, (640, 480))
        print(photo.shape)
        if status:
            s.send(np.ndarray.tobytes(photo))
        else:
            print("Not in frame")
    camera.release()


threading.Thread(target=receiver).start()
threading.Thread(target=sender).start()
```
